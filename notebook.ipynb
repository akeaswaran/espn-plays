```python
import pandas as pd
import numpy as np
import json

with open('plays.json') as data_file:    
    json_data = json.load(data_file)  
```

```python
plays_json = pd.json_normalize(json_data)
plays_json.head()
```

```python
plays_json.columns
```

```python
plays_json['yds_rushed'] = np.select(
    [
        plays_json['Time'].between(0, 30, inclusive=False), 
        plays_json['Time'].between(30, 60, inclusive=True)
    ], 
    [
        'Easy', # values here line up to conditions above 
        'Medium'
    ], 
    default='Unknown'
)
```
